# **Inferential Statistics.**


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

from pathlib import Path

filepath = Path('/content/drive/MyDrive/TechProAcademyProject/Cleaned_data.csv')

df= pd.read_pickle(filepath)
df.head()
```

## Hypothesis Testing for product category.
### Anova test, is a statistical test used to analyzed the difference between the means of more than two groups.
### One way Anova test tells if the dependent variable changes according to the leve of the independent variable.
- The null Hypothesis (H0): There SalesAmount has not difference between the product categories.
- The alternative Hypothesis (Ha): There is difference between the Product Categories.




```python
from scipy.stats import f_oneway

clothing_sales = df[df['ProductCategory'] == 'Clothing']['SalesAmount']
electronics_sales = df[df['ProductCategory'] == 'Electronics']['SalesAmount']
appliances_sales = df[df['ProductCategory']== 'Home Appliances']['SalesAmount']
```

```python
f_statistic, p_value = f_oneway(clothing_sales, electronics_sales, appliances_sales)
print("F-statistic:", f_statistic)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis. There is a significant difference between product categories.")
else:
    print("\nFail to reject the null hypothesis. There is no significant difference between product categories.")
```
